```python
from regex_engine import ResolvedIngredient
from regex_engine.application.dto.agent.parsed_ingredient import ParsedIngredient
from regex_engine.ports.regex_registry import RegexRegistryReader

string_data = """
2 jajka kurki zielononóżki
5 dużych łyżek twarogu wiejskiego
2 szklanki mleka
szczypta kurkumy
500 g makaronu pełnoziarnistego
200 mililitrów wody
sos rybny (lub sos sojowy)
"""

mapping: dict[str, ParsedIngredient] = {
    "2 jajka kurki zielononóżki": ParsedIngredient(
        raw_input="2 jajka kurki zielononóżki",
        amount=2,
        unit_size="",
        unit="",
        condition="",
        name="jajka kurki zielononóżki",
        extra=""
    ),
    "5 dużych łyżek twarogu wiejskiego": ParsedIngredient(
        raw_input="5 dużych łyżek twarogu wiejskiego",
        amount=5,
        unit_size="dużych",
        unit="łyżek",
        condition="",
        name="twarogu wiejskiego",
        extra=""
    ),
    "2 szklanki mleka": ParsedIngredient(
        raw_input="2 szklanki mleka",
        amount=2,
        unit_size="",
        unit="szklanki",
        condition="",
        name="mleka",
        extra=""
    ),
    "szczypta kurkumy": ParsedIngredient(
        raw_input="szczypta kurkumy",
        amount=1,
        unit_size="",
        unit="szczypta",
        condition="",
        name="kurkumy",
        extra=""
    ),
    "500 g makaronu pełnoziarnistego": ParsedIngredient(
        raw_input="500 g makaronu pełnoziarnistego",
        amount=500,
        unit_size="",
        unit="g",
        condition="",
        name="makaronu pełnoziarnistego",
        extra=""
    ),
    "200 mililitrów wody": ParsedIngredient(
        raw_input="200 mililitrów wody",
        amount=200,
        unit_size="",
        unit="mililitrów",
        condition="",
        name="wody",
        extra=""
    ),
    "sos rybny (lub sos sojowy)": ParsedIngredient(
        raw_input="sos rybny (lub sos sojowy)",
        amount=1,
        unit_size="",
        unit="",
        condition="",
        name="sos rybny",
        extra="lub sos sojowy"
    ),
}
```

```python
def print_registry(registry:RegexRegistryReader):
    print(f"Registry: {registry.kind.name}")
    for entry in registry.get_all():
        print(f"stem : {entry.stem}")
        print(f"variants: {entry.variants}")
        print(f"regex: {entry.pattern.pattern}")
        print("---------------------------------")
```

```python
def print_resolved_ingredients(ingredients:list[ResolvedIngredient]):
    for ingredient in ingredients:
        print(f"ingredient: {ingredient.name}")
        print(ingredient)
```

```python
from regex_engine.api import create_demo

regex_engine = create_demo(mapping)

await regex_engine.learn(string_data)

```
