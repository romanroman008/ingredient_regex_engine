```python
from regex_engine import AgentConfig, EngineConfig, create_engine
from regex_engine.config import DatabaseStorageConfig, FileStorageConfig

# file repository
file_storage  = FileStorageConfig(output_dir="output")

# SQLite repository
sqlite_storage  = DatabaseStorageConfig(
    database_url="sqlite:///regex_engine.db",
    create_schema=True,
)
# PostgreSQL repository example
postgres_storage_example = DatabaseStorageConfig(
    database_url="postgresql+psycopg://user:pass@localhost:5432/dbname",
    create_schema=True,
)

categorizer_config = AgentConfig(
    model="gpt-4o-mini",
    timeout=20,
    ensemble_size=5,
    max_retries=3
)
parser_config = AgentConfig(
    model="gpt-4o-mini",
    timeout=20,
    ensemble_size=5,
    max_retries=3
)

engine_config = EngineConfig(storage=sqlite_storage ,
                             parser=parser_config,
                             categorizer=categorizer_config
                             )

engine = await create_engine(engine_config)
```
